# Summary

Some preliminary observations on the results of the segmentwise analysis of the wideband Yassiada amphoras. The point clouds of 86-96, W8, X5, Y12, Y34, and the top four segments of Y12 (excluding the lower body / base, which is missing a large chunk) were compared with the point cloud of X12.

### A few notes:
- I've added CV as one of the statistical quantities in our segmentwise comparisons.
- All x and y axes are scaled and ticked consistently across all six notebooks, for ease of comparison.
- Amphora Y2 is missing a large chunk from segments 5 and 6, so those segments were not analyzed for Y2.

# Observations by statistical quantity

### Average point-to-nearest-neighbor distance (absolute NND)

Across all six pairs, rims (segment 1) consistently have the smallest average NND (all on the order of 2 mm).
- Perhaps this is because rims are smaller, easy to control, and less subject to deformation after being formed.
- The one exception is X12_X5, where the rims have the second smallest average NND. The bases (segment 6) seem to have very similar curvatures, and they have the smallest average NND, as shown in these side, top, and bottom views:

### Average point-to-nearest-neighbor distance scaled by segment's max diameter

Across all segments, the body segments have the smallest NNDs scaled by those segments' max diameters.

Necks have the highest NNDs in all cases except 86-96, where the base is significantly different from X12c relative to max base diameter. Rim NNDs are less than necks but more than bodies.

This all reflects what we saw with the main spiral comb group from the first time we did pairwise segmentwise analysis with five jars. For reference, links to those here:

Regarding the notable high values:

- Y34's neck is more different than usual from X12's neck, as shown above.

### CV of point-to-nearest-neighbor distance

We get high stdev and CV between two segments when some points are close together and other points are far from each other. The CV plot suggests that the neck is the segment where we see this behavior the most meaningfully. That is, in general, the neck segment NNDs have higher CV, and therefore dispersion around their average NNDs, incomparison with the CVs of the other segments.

# Other Observations

The smallness of many of these NNDs for small sections (e.g. consistenly around 2 mm for the rim pairs) reflects the fact that tighter alignment can be found for smaller sections. When we compare entire bodies, we find that it is impossible to align two entire bodies in such a way that on average, point-to-nearest-neighbor distance is less than 4 or 5 mm, since there is so much variation going on all throughout the large surfaces.
